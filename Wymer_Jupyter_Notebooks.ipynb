```python
#Web Scrapping with Jupyter

#Import tools needed
import urllib3
import certifi
import re
import pandas as pd


http = urllib3.PoolManager(ca_certs=certifi.where())

#Establish website and formats
url = "https://www.metacritic.com/browse/movies/score/metascore/year/filtered?year_selected=2006&sort=desc&view=detailed"
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:105.0) Gecko/20100101 Firefox/105.0"
response = http.request('GET', url, headers={"User-Agent": user_agent})
page_content = response.data.decode('utf-8')


#Build out regex setups
titles = r'class="title"><h3>(.*)<\/h3><\/a>'
release_dates = r'<span>(?P<ReleaseDate>.*\d)<\/span>'
summaries = r'<div class="summary">\s(.*)[\s\S]*?<\/div>'
meta_scores = r'<div class="metascore_w large movie positive">(.*)<\/div>\s*<\/a>\s*<\/div>\s*<div class="clamp-userscore">'
thumbnails = re.compile(r'"><img src="([h].*)" alt="', flags=re.M)

#Build out searches
m_titles = re.findall(titles, page_content)
m_release_dates = re.findall(release_dates, page_content)
m_summaries = re.findall(summaries, page_content)
m_meta_scores = re.findall(meta_scores, page_content)
m_thumbnails = thumbnails.findall(page_content)

#Checking lengths of each search run to make sure the searches functioned properly
#print(len(m_titles))
#print(len(m_release_dates))
#print(len(m_summaries))
#print(len(m_meta_scores))
#print(len(m_thumbnails))

#Build Pandas Dataframe to store outputs
movies = pd.DataFrame({
    'Title': m_titles,
    'Release Date': m_release_dates,
    'Summary': m_summaries,
    'Score': m_meta_scores,
    'Thumbnail': m_thumbnails
})

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None                       
                       ):
    display(movies)
```
